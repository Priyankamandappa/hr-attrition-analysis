```python
# Import all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
plt.style.use('seaborn-v0_8')

print("Libraries imported successfully!")
```

```python
# Load the HR Attrition dataset
df = pd.read_csv(r"C:\Users\PRIYANKA MANDAPPA\Downloads\HR attrition project\WA_Fn-UseC_-HR-Employee-Attrition.csv")

# First look at the data
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
# Check data types and missing values
print("Dataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Basic statistics of the dataset
print("Basic Statistics:")
print(df.describe())

print("\nAttrition Value Counts:")
print(df['Attrition'].value_counts())

print("\nAttrition Percentage:")
print(df['Attrition'].value_counts(normalize=True) * 100)
```

```python
# These columns have only 1 unique value — useless for analysis
cols_to_drop = ['EmployeeCount', 'Over18', 'StandardHours']

df = df.drop(columns=cols_to_drop)

print("Dropped unnecessary columns!")
print("New shape:", df.shape)
```

```python
# Convert Yes/No to 1/0 for analysis
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})

print("Attrition conversion done!")
print(df['Attrition_Flag'].value_counts())
```

```python
# See all categorical columns and their unique values
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    print(f"{col}: {df[col].unique()}")
```

```python
plt.figure(figsize=(8, 5))
colors = ['#2ecc71', '#e74c3c']
df['Attrition'].value_counts().plot(kind='bar', color=colors, edgecolor='black')
plt.title('Overall Employee Attrition Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Attrition')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig(os.path.join(images_path, '01_attrition_distribution.png'), dpi=150)
plt.show()
print("Chart saved!")
```

```python
dept_attrition = df.groupby('Department')['Attrition_Flag'].mean() * 100

plt.figure(figsize=(8, 5))
dept_attrition.sort_values(ascending=False).plot(
    kind='bar', color='#e74c3c', edgecolor='black')
plt.title('Attrition Rate by Department (%)', fontsize=14, fontweight='bold')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig(os.path.join(images_path, '02_attrition_by_department.png'), dpi=150)
plt.show()
print("Chart saved!")
```

```python
df['Age_Group'] = pd.cut(df['Age'], 
                          bins=[18, 25, 35, 45, 60], 
                          labels=['18-25', '26-35', '36-45', '46-60'])

age_attrition = df.groupby('Age_Group', observed=True)['Attrition_Flag'].mean() * 100

plt.figure(figsize=(8, 5))
age_attrition.plot(kind='bar', color='#3498db', edgecolor='black')
plt.title('Attrition Rate by Age Group (%)', fontsize=14, fontweight='bold')
plt.xlabel('Age Group')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig(os.path.join(images_path, '03_attrition_by_age.png'), dpi=150)
plt.show()
print("Chart saved!")
```

```python
role_attrition = df.groupby('JobRole')['Attrition_Flag'].mean() * 100

plt.figure(figsize=(12, 6))
role_attrition.sort_values(ascending=True).plot(
    kind='barh', color='#9b59b6', edgecolor='black')
plt.title('Attrition Rate by Job Role (%)', fontsize=14, fontweight='bold')
plt.xlabel('Attrition Rate (%)')
plt.ylabel('Job Role')
plt.tight_layout()
plt.savefig(os.path.join(images_path, '04_attrition_by_jobrole.png'), dpi=150)
plt.show()
print("Chart saved!")
```

```python
plt.figure(figsize=(10, 5))
df.boxplot(column='MonthlyIncome', by='Attrition', 
           patch_artist=True,
           boxprops=dict(facecolor='#3498db', color='black'),
           medianprops=dict(color='red', linewidth=2))
plt.title('Monthly Income vs Attrition', fontsize=14, fontweight='bold')
plt.suptitle('')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.tight_layout()
plt.savefig(os.path.join(images_path, '05_income_vs_attrition.png'), dpi=150)
plt.show()
print("Chart saved!")
```

```python
wlb_attrition = df.groupby('WorkLifeBalance')['Attrition_Flag'].mean() * 100

plt.figure(figsize=(8, 5))
wlb_attrition.plot(kind='bar', color='#e67e22', edgecolor='black')
plt.title('Attrition Rate by Work-Life Balance Score', fontsize=14, fontweight='bold')
plt.xlabel('Work-Life Balance (1=Bad, 4=Best)')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig(os.path.join(images_path, '06_worklifebalance_attrition.png'), dpi=150)
plt.show()
print("Chart saved!")
```

```python
ot_attrition = df.groupby('OverTime')['Attrition_Flag'].mean() * 100

plt.figure(figsize=(7, 5))
ot_attrition.plot(kind='bar', color=['#2ecc71', '#e74c3c'], edgecolor='black')
plt.title('Attrition Rate by Overtime (Yes/No)', fontsize=14, fontweight='bold')
plt.xlabel('Overtime')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig(os.path.join(images_path, '07_overtime_attrition.png'), dpi=150)
plt.show()
print("Chart saved!")
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(16, 10))
correlation = numeric_df.corr()

sns.heatmap(correlation, 
            annot=True, 
            fmt='.2f',
            cmap='RdYlGn',
            center=0,
            linewidths=0.5,
            annot_kws={"size": 7})

plt.title('Correlation Heatmap — HR Attrition Dataset', 
          fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig(os.path.join(images_path, '08_correlation_heatmap.png'), dpi=150)
plt.show()
print("Chart saved!")
```

```python
print("=" * 55)
print("        HR ATTRITION ANALYSIS — KEY INSIGHTS")
print("=" * 55)

print(f"\n📊 Total Employees: {len(df)}")
print(f"🚨 Overall Attrition Rate: {df['Attrition_Flag'].mean()*100:.1f}%")
print(f"👥 Employees Who Left: {df['Attrition_Flag'].sum()}")

print("\n🏢 Attrition by Department:")
print(df.groupby('Department')['Attrition_Flag'].mean().mul(100).round(1).to_string())

print("\n💼 Top 3 Job Roles with Highest Attrition:")
print(df.groupby('JobRole')['Attrition_Flag'].mean().mul(100).round(1)
      .sort_values(ascending=False).head(3).to_string())

print("\n⏰ Attrition by Overtime:")
print(df.groupby('OverTime')['Attrition_Flag'].mean().mul(100).round(1).to_string())

print("\n👶 Attrition by Age Group:")
print(df.groupby('Age_Group', observed=True)['Attrition_Flag'].mean()
      .mul(100).round(1).to_string())

print("\n💰 Average Income — Stayed vs Left:")
print(df.groupby('Attrition')['MonthlyIncome'].mean().round(0).to_string())

print("\n" + "=" * 55)
print("✅ Analysis Complete!")
print("=" * 55)
```
